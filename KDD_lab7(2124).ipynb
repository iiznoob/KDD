```python
# Import the required libraries
import pandas as pd
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.preprocessing import TransactionEncoder
```

```python
# Load the dataset
data = pd.read_csv('/content/Market_Basket_Optimisation.csv')
data.head(5)
```

```python
# Preprocess the data into transaction format
transactions = []
for i in range(len(data)):
    transactions.append([str(data.values[i,j]) for j in range(len(data.columns))])
# transactions
```

```python
# Encode the transaction data using TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit_transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
# Use FP-growth to generate frequent itemsets
frequent_itemsets = fpgrowth(df, min_support=0.05, use_colnames=True)
```

```python
# Print the frequent itemsets
print(frequent_itemsets)
```
