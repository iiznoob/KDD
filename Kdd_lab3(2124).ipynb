```python
 # Importing Required libraries import pandas as pd 
import numpy as np
import pandas as pd
```

```python
Url = "https://raw.githubusercontent.com/shrikant-temburwar/Loan-Prediction-Dataset/master/train.csv"
df = pd.read_csv(Url)
df
```

```python
all_data_na = (df.isnull().sum() / len(df)) * 100 
all_data_na = all_data_na.drop(all_data_na[all_data_na == 0].index).sort_values()
missing_data =pd.DataFrame({'Missing Ratio (in %)':all_data_na}) 
missing_data.head(20)
```

```python
df = df['LoanAmount'].fillna(df['LoanAmount'].mean())
```

```python
# Functions for Normalization
# min-max normalization function
def minmaxx(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result

# z-score normalization function
def z_score_normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        mean_value = df[feature_name].mean()
        std_value = df[feature_name].std()
        result[feature_name] = (df[feature_name] - mean_value) / std_value
    return result

# log-scale normalization function
def log_scale_normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        result[feature_name] = np.log(df[feature_name])
    return result

# decimal scaling normalization function
def decimal_scale_normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        j = 0
        while max_value > 1:
            max_value /= 10
            j += 1
        result[feature_name] = df[feature_name] / (10 ** j)
    return result
```

```python
# Select the desired columns
data_fornormal = df[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term']]

# Apply the minmaxx function
minmax_norm_data = minmaxx(data_fornormal)
minmax_norm_data
```

```python
# Apply Z_score_norm
z_scorenorm = z_score_normalize(data_fornormal)
z_scorenorm
```

```python
# Apply decimal Sclae
decimal_norm = decimal_scale_normalize(data_fornormal)
decimal_norm
```

```python
# Functions for Standardization
# mean_standardization function
def mean_standardize(df):
    result = df.copy()
    for feature_name in df.columns:
        mean_value = df[feature_name].mean()
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = (df[feature_name] - mean_value) / (max_value - min_value)
    return result

# scaling to unit length function
def scale_to_unit_length(df):
    result = df.copy()
    for feature_name in df.columns:
        norm_value = np.linalg.norm(df[feature_name])
        result[feature_name] = df[feature_name] / norm_value
    return result
```

```python
data_scale = df[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term']]

# Apply Mean_standard
mean_stand = mean_standardize(data_scale)
mean_stand
```

```python
# Apply Scale to unit
mean_stu = scale_to_unit_length(data_scale)
mean_stu
```

```python
# Using Library
from sklearn.preprocessing import StandardScaler, FunctionTransformer
# Z-score normalization (standardization) using StandardScaler
scaler = StandardScaler()
z_score_normalized_data = scaler.fit_transform(data_scale)
print('Z-score normalized data:')
print(z_score_normalized_data)

```

```python
# Log scaling using FunctionTransformer
transformer = FunctionTransformer(np.log1p)
log_scaled_data = transformer.transform(data_scale)
print('Log scaled data:')
print(log_scaled_data)
```
